```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Image
```

```python
image_path = "D://Pics//Camera Roll//IMG_20201215_121027_567.jpg"
original_image = cv2.imread(image_path)

# Check if the image was loaded successfully
if original_image is not None:
    # Display the original image
    plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for proper display
    plt.title("Original Image")
    plt.axis('off')  # Optional: Turn off axis labels
    plt.show()
else:
    print("Image not found or could not be loaded.")
```

```python
gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
```

```python
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
```

```python
edges = cv2.Canny(blurred_image, threshold1 = 25, threshold2 = 90)
```

```python
display(Image(data = cv2.imencode('.png', edges)[1].tobytes(), format = 'png'))
```
