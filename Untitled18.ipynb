# 2. Region based Image Segmentation

```python
# Read the image
image = cv2.imread("C:\\Users\\akhil\\OneDrive\\Desktop\\1.jpeg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding to create a binary image
ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Apply morphological operations to remove noise and improve segmentation
kernel = np.ones((3, 3), np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=2)

# Finding sure background area
sure_bg = cv2.dilate(opening, kernel, iterations=3)

# Finding sure foreground area
dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
ret, sure_fg = cv2.threshold(dist_transform, 0.7 * dist_transform.max(), 255, 0)

# Finding unknown region
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg, sure_fg)

# Label markers for watershed
ret, markers = cv2.connectedComponents(sure_fg)
markers = markers + 1
markers[unknown == 255] = 0

# Apply watershed algorithm
markers = cv2.watershed(image, markers)
image[markers == -1] = [0, 0, 255]  # Mark segmented regions with red color

# Display the result using Matplotlib in the notebook
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Turn off axis labels
plt.show()
```

# 3. Region based Image Segmentation

```python
def region_growing(image, seed):
    height, width = image.shape
    visited = np.zeros_like(image)
    segmented = np.zeros_like(image)
    queue = []

    queue.append(seed)
    visited[seed] = 1
    region_sum = 0
    region_count = 0

    while queue:
        pixel = queue.pop(0)
        x, y = pixel

        if x > 0 and x < height - 1 and y > 0 and y < width - 1:
            for i in range(-1, 2):
                for j in range(-1, 2):
                    if visited[x + i, y + j] == 0 and abs(int(image[x, y]) - int(image[x + i, y + j])) < 30:
                        queue.append((x + i, y + j))
                        visited[x + i, y + j] = 1
                        segmented[x + i, y + j] = image[x + i, y + j]
                        region_sum += int(image[x + i, y + j])
                        region_count += 1

    return segmented, region_sum, region_count

# Read the image
image = cv2.imread("C:\\Users\\akhil\\OneDrive\\Desktop\\1.jpeg", cv2.IMREAD_GRAYSCALE)

# Define a seed point (pixel coordinates) for the region growing
seed_point = (100, 100)

# Perform region growing segmentation
segmented_image, region_sum, region_count = region_growing(image, seed_point)

# Display the segmented image and print region info using Matplotlib
plt.figure(figsize=(10, 6))

# Display the original image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Display the segmented image
plt.subplot(1, 2, 2)
plt.imshow(segmented_image, cmap='gray')
plt.title('Segmented Image')
plt.axis('off')

# Print region information
plt.suptitle(f"Region Sum: {region_sum}\nRegion Count: {region_count}", fontsize=14)
plt.show()
```
